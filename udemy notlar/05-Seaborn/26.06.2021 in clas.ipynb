```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age,salary)
```

```python
fig,ax=plt.subplots()
ax.plot(age,salary)
```

```python
plt.plot(age,salary,label="1")
plt.plot(age,salary_2,label="2")
plt.legend()        # label verirsek etiket oluyor legend
```

```python
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.plot(age, salary)
plt.subplot(1,2,2)
plt.plot(age, salary_2)
plt.tight_layout()   # 2 si burda iyi görünmedi
```

```python
fig,ax1=plt.subplots(nrows=2,ncols=2)
ax1[0, 0].plot(age, salary)
ax1[0, 1].plot(age, salary_2)
plt.tight_layout()
```

```python
fig, ax1 = plt.subplots(nrows=2, ncols=2)
ax1[0, 0].plot(age, salary)
ax1[0, 1].plot(age, salary_2)
plt.tight_layout()
```

```python
def multi_graf(rows,cols):
    fig, ax = plt.subplots(nrows=rows, ncols=cols)
    for i in range(cols):
        for j in range(rows):
            ax[i,j].plot(age,salary)
            plt.tight_layout()
multi_graf(2,2)
```
