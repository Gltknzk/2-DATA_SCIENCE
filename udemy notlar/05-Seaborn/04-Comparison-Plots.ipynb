# Comparison Plots with pairplot() and jointplot()

## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## The Data

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

## jointplot

```python
sns.jointplot(x="math score",y="reading score",data=df)
```

```python
sns.jointplot(x='math score',y='reading score',data=df)
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind="hex")
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind="resid")# | "hist" | "hex" | "reg" | "resid" }hex')
```

```python
sns.jointplot(x='math score',y='reading score',data=df,kind='kde',alpha=0.8,shade=True,hue="gender")
```

## pairplot

```python
sns.pairplot(data=df)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='gender',palette='viridis')
```

```python
sns.pairplot(df,hue='gender',palette='viridis',corner=True)
```

```python
sns.pairplot(df,hue='gender',palette='viridis',diag_kind='hist')
```
