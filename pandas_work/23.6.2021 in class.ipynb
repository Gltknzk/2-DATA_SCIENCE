.merge()

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how="inner", on="key")  # kesişenler aynı sutune sahip df aldık
```

```python
pd.merge(left,right,how="outer", on="key")
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                     'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how="inner", on="key")  # k0 k1 k2 ortak  yanyana aynı sütünlü ortak sutunlar üstüne merge ettik
                                            #kesişenler aynı sutune sahip df aldık  on key
```

```python
pd.merge(left,right,how="outer", on="key")  # k0 k1 k2 ortak  yanyana aynı sütünlü ortak sutunlar üstüne merge ettik
                                            #kesişenler aynı sutune sahip df aldık  on key
                                           # outer hepsi geldi.
```

```python
pd.merge(left,right, how="left", on="key")  # sol df göre 
```

```python
pd.merge(left,right, how="right", on="key")  # sağ df göre orda k3 vardı c ve d sütünları için değer verdi
```

```python
left1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
right1 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left1,right1,how="outer", on=["key1","key2"])  # how inner default on liste içinde yazdık
                                                        # a2 b2 tekrar etti.
```

```python
df11 = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x', 'z'],
                    'lvalue': [2, 3, 5, 7, 0]})
df12 = pd.DataFrame({'rkey': ['a', 'x', 'z', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
df11
```

```python
df12
```

```python
pd.merge(df11,df12,left_on= "lkey",right_on= "rkey", how="outer") #left te lkeyi right rkey i al. 
```

```python
pd.merge(df11,df12,left_on= "lkey",right_on= "rkey", how="left") #left ti aldı leftin üstüne df11 di
                                                       # eksik olursa tekra eder, hiç yoksa nan
```

```python
pd.merge(df11,df12,left_on= "lkey",right_on= "rkey", how="right") #rght satır sayısı arttı r de 4 dü leftte 5 di sonuç altı satır oldu
```

# .join() index bazında işlem yapar

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index = ['K0', 'K1', 'K2']) 
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index = ['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
left.join(right)  # lefti baz alıp üstüne joın indexx üzerinden
```

```python
left.join(right,how="outer") # tüm indexler baz alınsn hepsini getir değerleri yaz diğer nan
```

```python
left.join(right,how="inner") # joıin defaultu leftti inner kesişenleri getir.
```

```python
left.join(right,how="right") # right baz al 
```
