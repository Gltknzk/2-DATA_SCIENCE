```python
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
 df["t_quality"]=np.vectorize(quality)(df["total_bill"],df["tip"]) # hızlı numpy da total_bill ve tip üzerinde
```
