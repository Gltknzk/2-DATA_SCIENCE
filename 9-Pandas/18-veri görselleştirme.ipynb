```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")  # sea born küt. data yükle
df.head()
```

# BARPLOT en çok kullanılna sözel ve sayısal kütüphane  seabornde...

```python
sns.barplot(x="sex",y="survived",data=df);    # cinsiyete göre hayatta kalma oranını grafikle
```

```python
sns.barplot(x="survived",y="sex",data=df);
```

```python
sns.barplot(x="sex",y="survived",hue="pclass",data=df);
# clasa göre ayar ekince sutünü böldü
```

```python
sns.barplot(x="sex",y="survived",data=df);
```

```python
tips=sns.load_dataset("tips")
df=tips.copy()  # yedek aldık
df.head()
```

```python
df.isnull().sum()     #boş veri varmı yokkk
```

```python
df["sex"].value_counts()  #cinsiyet toplamı
```

```python
df["smoker"].value_counts()
```

```python
df["day"].value_counts()  #sadece count olursa sayar verir
```

```python
sns.barplot(x="sex",y="total_bill",data=df);
# erkekler daha fazla bahşiş vermiş (kıyas yaptı)
```

```python
sns.barplot(x="sex",y="total_bill",hue="time",data=df); # akşam  ve erkekler daha harcamış garson akşam iyiler
```

```python
sns.barplot(x="sex",y="tip",data=df);  #kim ne ödemiş
```

```python
sns.barplot(x="sex", y="tip", hue="smoker",data=df);  #sigara içenlere göre ayırdı
```

```python
df.head()
```

```python
sns.barplot(x="day",y="total_bill",hue="time",data=df);   # çok güzel 
```

### boxplot ikinci kullanılan  kutu gibi

```python
sns.boxplot(x="sex",y="total_bill",data=df);  # burda erkekler daha fazla harcama yapmış
                                             #  dağılım dışına çıkan veriler sapmalar
                                            #   kutunun üstü en çok değer ltı en alt değer
                                           #    çizgi medyan değer
```

```python
sns.boxplot(x="sex",y="total_bill",hue="time",data=df);
```

```python
### vionplot
```

```python
sns.catplot(x="sex",y="tip",kind="violin",data=df); # burda kind violin kütüphanesinden al göbek genişse orda yoğunluk
```

```python
sns.catplot(x="sex",y="tip",data=df); # burda noktaladı yığılma yaptı
```

```python
sns.catplot(x="sex",y="total_bill",kind="violin", hue="day",data=df); # violinle günlere göre byan erkek harcama
```

```python
sns.catplot(x="sex",y="total_bill",hue="day",data=df); # catplota göre bayan erkek harcama
```

### scatterplot   iki sayısal değeri aynı anda kıyaslama:

```python
sns.scatterplot(x="total_bill",y="tip",data=df);  # hesap artınca bahşiş artmış
```
