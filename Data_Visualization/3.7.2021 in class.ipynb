```python
fig,ax_tip=plt.subplots(figsize=(12,6))
ax_tip=sns.countplot(x="day",data=tips)
#sns.countplot(x="day",data=tips,ax_tip=)# sns de coutplot saydÄ± hangigÃ¼n ne kadr alÄ±ÅŸveriÅŸ

for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1))
    ax_tip.annotate("%"+str(round((p.get_height()/tips.day.count()*100),2)), (p.get_x()+0.3, p.get_height()-5));
```
