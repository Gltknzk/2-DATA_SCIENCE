# jointplot    t yi unutma joint

```python
sns.jointplot(data=tips,x="total_bill",y="tip", kind="scatter")   # scatter plot ve histogram birleÅŸti.
                                                                #  { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
```

```python
sns.jointplot(data=tips,x="total_bill",y="tip", kind="hist")   # scatter plot ve histogram birleÅŸti.
                                                                #  { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
```
