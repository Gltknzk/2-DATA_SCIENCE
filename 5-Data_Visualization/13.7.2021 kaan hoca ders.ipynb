```python
def in_thousand (inst):
    if inst == '500.0 k':  # k olanları m yaptık
        return '0.5 M' 
    elif inst == '100.0 k': # k olanları m yaptık
        return '0.1 M'
    else:
        return inst
df['installs']= df['installs'].apply(in_thousand) # aplly ile uyguladık

df['installs']= df['installs'].str.replace( 'M', '').str.strip().astype('float') # boşlukları düşürdük

df= df.rename(columns={'installs': 'installs_in_million'})
df['installs_in_million'].value_counts()
```
