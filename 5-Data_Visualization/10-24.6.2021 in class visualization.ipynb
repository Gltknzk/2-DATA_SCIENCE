```python
x = np.arange(1,16)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4)

ax[1].plot(x,x**2,x,x**3,x,x**4)
ax[1].set_xlim([2,4])     # zoom yapma iÅi  limit iÅi
ax[1].set_ylim([5,100])   # zoom yapma iÅi saÄda limitleyerek getirdi
```

```python
x = np.arange(1,16)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4)

ax[1].plot(x,x**2,x,x**3,x,x**4)
ax[1].set_xlim([2,4])   # zoom yapma iÅi  limit iÅi
ax[1].set_ylim([5,100])   # zoom yapma iÅi

ax[0].axvline(x=3,ls="--")  # zoom yaptÄ±m saÄda da gÃ¶sterdim
ax[0].axhline(y=50,ls="--")

ax[1].axvline(x=3,ls="--")  # zoom yaptÄ±Äm yeri saÄda da gÃ¶sterdim kÄ±rÄ±lmalarÄ± gÃ¶r
ax[1].axhline(y=50,ls="--")
```
