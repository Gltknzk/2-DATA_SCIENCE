```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


%matplotlib inline  
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
# functional olarak (fonksiyon çağırarak. fonksiyonun içine değişkeni/değişkenleri yazarak)

plt.plot(salary)
plt.show()
```

```python
# functional

plt.plot(age, salary)
plt.show()

# şimdi age değişkenleri ile birleşti.
```

```python
# iki çizgi çıksın istersem:
# ikinci çizgi için farklı bir değişken (salary_2) kullanarak y eksinini kaydırıyorum

plt.plot(age, salary)
plt.plot(age, salary_2)#ikinci plotta y eksinine farklı değer girdiğim için ikinci çizgiyi çizgi
```

```python
# legend kullanırsak:
# legend'de label ları ya plt.plot() fonksiyonunun içinde tanımlıyoruz.
 
plt.plot(age, salary, label="1")
plt.plot(age, salary_2, label="2")
plt.legend()

```

```python
# veya legend() fonksiyonunun içinde de tanımlayabiliriz.
 
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.legend(["1","2"])
```

```python
plt.plot(age, salary, age, salary_2) 

# iki farklı axes'ı (age-salary ve age-salary_2) sırayla fonksiyon içine soktum. iki çizgili şekil elde ettim.
```

```python
# fonksiyonda 2 axes kullandığımda legend'in label'larını fonksiyon içinde değil, legend() fonksiyonunun içinde tanımlıyorum!!!

plt.plot(age, salary, age, salary_2)
plt.legend(["1", "2"])
```

```python
# object oriented olarak:

# önce object'leri oluşturuyoruz ( tanımlıyoruz )

# sonra .plot metodu ile verilerimi giriyoruz

fig, ax = plt.subplots()
ax.plot(age, salary)
```

```python
# 1. figure size ını genişleterek düzeltebiliriz.

plt.figure(figsize=(16,4))

plt.subplot(1,2,1)
plt.plot(age, salary)

plt.subplot(1,2,2)
plt.plot(age, salary_2)
```

```python
# 2. plt.tight_layout() ile düzelttik.

plt.subplot(1,2,1)
plt.plot(age, salary)

plt.subplot(1,2,2)
plt.plot(age, salary_2)

plt.tight_layout()
```

```python


fig, ax1 = plt.subplots(nrows=1, ncols=2)
ax1[0].plot(age, salary)
ax1[1].plot(age, salary_2)
plt.tight_layout()
```

```python
fig, ax1 = plt.subplots(nrows=2, ncols=2)
ax1[0,0].plot(age, salary)  # age-salary'i 0-0 indekse çizdi
ax1[0,1].plot(age, salary_2) # age-salary_2'yi 0-1 indekse çizdi
plt.tight_layout()  

# geri kalan 2 indeks için veri tanımlanmadığı için boş olarak çizdi
```

```python
def multi_graf(rows, cols):
    fig, ax = plt.subplots(nrows=rows, ncols=cols)
    for i in range(rows):
        for j in range(cols):
            ax[i,j].plot(age, salary)
            plt.tight_layout()
multi_graf(2,2)   # bravo def i anladım
```

```python
# şimdi bunu bar plotun içine atalım ve şekli görelim.
```

```python
day = tips.day.unique()
total_bill = tips.groupby("day")["total_bill"].sum()

fig, ax = plt.subplots()
ax.bar(day, total_bill)
plt.show()

# day'i belirlerken unique dediğimi için gün sıralaması yanlış oldu.
```

```python
# gün gruplaması yaparak day-total_bill plotuna bakalım

day = tips.groupby("day")["total_bill"].sum().index
total_bill = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(day,total_bill)
plt.show()
```

```python
# gün gruplaması yaparak day-tip plotuna bakalım

day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(day,tip)
plt.show()
```

```python
# cinsiyet gruplaması yaparak sex-tips plotuna bakalım

sex = tips.groupby("sex")["total_bill"].sum().index
tip = tips.groupby("sex")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(sex,tip)
plt.show()
```

```python
# iki axes'i birden görüntülesin istiyorum:

day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
total = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(day, total, label="total" )
ax.bar(day, tip, label="tip" )
plt.legend()
plt.show()
```

```python
# tip sütunlarını daraltınca total ondan geniş kalacağı için daha güzel bir görünüm ortaya çıktı.

day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
total = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()
ax.bar(day, tip, label="tip", align="edge")
ax.bar(day, total, label="total",width=0.6, align="edge")
plt.legend()
plt.show()
```

# Box Plot

```python
# az önceki x değişkenine tanımladığım datamın box plotunu görelim:

x = np.random.normal(170,10,250)
plt.boxplot(x)
plt.show()
```

```python
# birden fazla box plotu görmek istersem:

box = np.array(np.random.rand(10,4))
plt.boxplot(box, labels = ["A", "B", "C", "D"])  
plt.show()

# box, her defasında random olarak farklı değerler üretecektir. her harfi temsil eden boxplotlar farklı değerlere sahip olacaktır.
```

```python
box = np.array(np.random.rand(10,4))
plt.boxplot(box, labels=["a","b","c","d"], vert= False)
plt.show()

# vert=False yaptım. horizontal oldu.
```

```python
# tips datasetimdeki total_bill sütununun boxplotunu görelim:

plt.figure(figsize=(16,6))
plt.boxplot(tips.total_bill, vert=False)
plt.show()
```
