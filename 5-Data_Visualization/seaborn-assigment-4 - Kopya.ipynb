## 2: Plot 12 most dangerous cities

```python
df.head()
```

```python
df.isnull().sum()
```

```python
most_danger_city = df.city.value_counts()[:12]
```

```python
most_danger_city
```

```python
most_danger_city.plot.bar()
```

```python
x = most_danger_city.index
x
```

```python
y = most_danger_city.values
y
```

```python
plt.figure(figsize=(12,6),dpi = 150)
sns.barplot(x,y, palette='viridis')
plt.xticks(rotation = 60);
for index , value in enumerate(y):
    plt.text(index, value+0.1, str(value), ha = 'center')
plt.title('Most Dangerous City',color = 'orange');
```

```python
most_danger_city.index
```

```python
sns.barplot(x = df.city.value_counts()[:12].index, y = df.city.value_counts()[:12].values)
sns.barplot(x=df.city.value_counts()[:12].index,y=df.city.value_counts()[:12].values)
plt.xticks(rotation=60);
```

```python
most_danger_city.index
```

```python
plt.figure(figsize=(16,6))
sns.countplot(data = df , x = 'city', order = most_danger_city.index);
sns.countplot(data=df,x="city",order= most_danger_city.index)
plt.xticks(rotation = 60)
plt.show()
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
mds = df.state.value_counts()[:20] # mds = shorteness of most dangerous state
```

```python
mds
```

```python
plt.figure(figsize=(12,6))
sns.barplot(mds.index, mds.values);
plt.xticks(rotation = 60);
plt.title('Most Dangerous States');
plt.legend
```

```python
# with seaborn barplot
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data = df , x = 'state', order=mds.index)
plt.title('Most Dangerous States');
```

## 5: Plot number of Threat Types

```python
threat_level = df.threat_level.value_counts()
threat_level
```

```python
threat_level.plot.bar()
plt.xticks(rotation=60)
for index , value in enumerate(threat_level.values):
    plt.text(index, value+0.1, str(value), ha = 'center')
```

```python
sns.barplot(x = threat_level.index, y = threat_level.values)
```
